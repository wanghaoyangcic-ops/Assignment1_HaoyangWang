```python
# Final working solution
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        if len(s) != len(t):
            return False # Return False directly if results are not the same
        count_s = {}# Create a dictionarie to count s
        count_t = {}# Create a dictionarie to count t
        for ch in s:# How many s
            if ch in count_s:
                count_s[ch] = count_s[ch] + 1
            else:
                count_s[ch] = 1
        for ch in t:# How many t
            if ch in count_t:
                count_t[ch] = count_t[ch] + 1
            else:
                count_t[ch] = 1
        for key in count_s:# Compare 2 dics
            if key not in count_t:
                return False
            if count_s[key] != count_t[key]:
                return False
        for key in count_t: # iF t contains extra index
            if key not in count_s:
                return False
        return True

# I used ChatGPT (Level 4) to learn the usage of char and how to implement it in this specific problem.
```
