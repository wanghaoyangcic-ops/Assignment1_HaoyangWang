```python
def reverse(x):
    INT_MIN = -2**31
    INT_MAX = 2**31 - 1
    if x < 0:
        sign = -1
        x = -x
    else:
        sign = 1
    rev = 0
    while x > 0:
        digit = x % 10
        rev = rev * 10 + digit
        x = x // 10
    rev = rev * sign
    if rev < INT_MIN or rev > INT_MAX:
        return 0
    else:
        return rev
```
